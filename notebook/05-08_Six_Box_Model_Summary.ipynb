## 1. 6-box model の全体像 / Overview of 6-box model

```text
N → M → D → S → P → L → N
```

この循環により、表層・中層・深層が結ばれます。

This circulation links the surface, intermediate, and deep ocean.


````python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(figsize=(8,5))
pos={'P':(0.1,0.7),'S':(0.3,0.7),'L':(0.5,0.7),'N':(0.8,0.7),'M':(0.5,0.42),'D':(0.5,0.15)}
for k,(x,y) in pos.items():
    ax.text(x,y,k,ha='center',va='center',bbox=dict(boxstyle='round',fc='white'))
path=[('N','M'),('M','D'),('D','S'),('S','P'),('P','L'),('L','N')]
for a,b in path:
    ax.annotate('',xy=pos[b],xytext=pos[a],arrowprops=dict(arrowstyle='->'))
ax.set_xlim(0,1);ax.set_ylim(0,1);ax.axis('off')
plt.show()
````
