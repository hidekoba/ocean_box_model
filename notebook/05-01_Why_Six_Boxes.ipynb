## 今日の目的 / Goals

ここまで、4-box モデルでは次のような海洋を考えました。  
So far, in the 4-box model, we considered an ocean like this:

```text
S: Subpolar surface / 南大洋表層
L: Low-latitude surface   / 低緯度表層
N: North Atlantic surface / 北大西洋表層
D: Deep ocean             / 深層
```

4-box モデルは、南大洋と北大西洋の役割を分けるうえで有用です。  
The 4-box model is useful for separating the roles of the Southern Ocean and the North Atlantic.

しかし、より現実的に海洋炭素循環を考えるには、まだ足りない点があります。  
However, to think more realistically about ocean carbon cycling, some important pieces are still missing.

この Notebook では、4-box から自然に拡張して、Toggweiler 型の 6-box モデル

```text
P S L N M D
```

を導入します。  
In this notebook, we naturally extend the 4-box model and introduce the Toggweiler-style six-box model:

```text
P S L N M D
```

今日の主題はこれです。  
The main theme is:

> **4-box では見えなかった極域表層 P と中層 M を追加し、海洋炭素循環をより立体的に見る。**  
> **Add Polar surface P and mid-depth M, which were not visible in the 4-box model, and view ocean carbon cycling more three-dimensionally.**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (7, 4)
````

## 1. Question: 4-box では何が足りないのか / What is missing in 4-box?

4-box モデルは、次のような問いには答えやすいです。  
The 4-box model is good for questions like:

```text
南大洋ベンチレーションが弱くなると大気 CO2 はどうなるか？
北大西洋沈み込みは深層炭素にどう効くか？
生物ポンプは大気 CO2 を下げるか？
```

```text
What happens to atmospheric CO2 if Southern Ocean ventilation weakens?
How does North Atlantic sinking affect deep carbon?
Does the biological pump lower atmospheric CO2?
```

しかし、次の問いには答えにくいです。  
But it is not good enough for questions like:

```text
極域表層は低緯度表層と同じ扱いでよいのか？
中層水と深層水の違いは重要ではないのか？
北大西洋で沈み込んだ水は、どの深さを通って戻るのか？
南大洋で湧昇する水は、中層から来るのか、深層から来るのか？
```

```text
Can Polar surface water be treated the same as low-latitude surface water?
Is the difference between mid-depth and deep water important?
At what depth does water sinking in the North Atlantic return?
Does upwelled Southern Ocean water come from mid-depth or deep water?
```

このため、4-box から 6-box へ進みます。  
For these reasons, we move from 4 boxes to 6 boxes.

## 2. Prediction: 何を追加すべきか / What should we add?

4-box から 6-box へ拡張するとき、追加したい要素は 2 つです。  
When extending from 4-box to 6-box, we want to add two elements.

### 1. 極域表層 P / Polar surface P

4-box では、低緯度表層をまとめて L として扱っていました。  
In the 4-box model, low-latitude surface waters were grouped into L.

しかし、極域表層は広く、炭素循環や水塊構造を考えるうえで重要です。  
However, the Polar surface is large and important for carbon cycling and water-mass structure.

そのため、P を独立させます。  
Therefore, we separate P.

### 2. 中層 M / Mid-depth M

4-box では、内部海洋を D だけで表していました。  
In the 4-box model, the ocean interior was represented only by D.

しかし、現実には中層水と深層水は同じではありません。  
But in reality, mid-depth water and deep water are not the same.

そのため、中層 M と深層 D を分けます。  
Therefore, we separate mid-depth M and deep D.

## 3. Toggweiler 型 6-box: PSLNMD

Toggweiler 型の 6-box モデルでは、海洋を次の 6 つに分けます。  
In the Toggweiler-style six-box model, the ocean is divided into six boxes:

| Box | 日本語 | English | 層 |
|---|---|---|---|
| P | 極域表層 | Polar surface | 表層 |
| S | 南大洋表層 | Subpolar surface | 表層 |
| L | 低緯度表層 | Low-latitude surface | 表層 |
| N | 北大西洋表層 | North Atlantic surface | 表層 |
| M | 中層 | Mid-depth / intermediate ocean | 内部 |
| D | 深層 | Deep ocean | 内部 |

つまり、

```text
P, S, L, N = surface boxes
M, D       = interior boxes
```

です。

重要なのは、6-box にした目的が単に箱を増やすことではない点です。  
The important point is that the purpose of six boxes is not simply to increase the number of boxes.

目的は、

```text
表層の地理的違い
内部の鉛直構造
表層と内部のつながり
```

を表現することです。  
The purpose is to represent:

```text
geographic differences in the surface ocean
vertical structure of the interior ocean
connections between surface and interior
```

````python
boxes = ["P", "S", "L", "N", "M", "D"]

box_info = pd.DataFrame({
    "Box": boxes,
    "Japanese": ["極域表層", "南大洋表層", "低緯度表層", "北大西洋表層", "中層", "深層"],
    "English": ["Polar surface", "Subpolar surface", "Low-latitude surface",
                "North Atlantic surface", "Mid-depth / intermediate", "Deep ocean"],
    "Layer": ["surface", "surface", "surface", "surface", "interior", "interior"],
})
box_info
````

## 4. Discussion: 4-box からどう拡張されたか / How is 4-box extended?

4-box の基本構造は、

```text
S, L, N, D
```

でした。  
The basic 4-box structure was:

```text
S, L, N, D
```

6-box では、次のように拡張します。  
In the six-box model, we extend it as follows:

```text
L の一部を P として分ける
D を M と D に分ける
```

```text
separate part of L as P
divide D into M and D
```

つまり、

```text
4-box: S, L, N, D

6-box: P, S, L, N, M, D
```

です。

この拡張により、次のことを考えられるようになります。  
This extension allows us to consider:

```text
Polar surface vs low-latitude surface
mid-depth vs deep ocean
surface circulation pathway
interior return pathway
```

```text
極域表層と低緯度表層の違い
中層と深層の違い
表層循環経路
内部の戻り経路
```

## 5. Experiment: PSLNMD を模式図にする / Draw the PSLNMD schematic

まず、PSLNMD の構造を図にします。  
First, we draw the PSLNMD structure.

これは厳密な海洋地図ではなく、概念図です。  
This is not an exact ocean map, but a conceptual diagram.

目的は、箱のつながりを直感的に理解することです。  
The purpose is to intuitively understand how boxes are connected.

````python
fig, ax = plt.subplots(figsize=(8, 5))

pos = {
    "P": (0.12, 0.70),
    "S": (0.36, 0.70),
    "L": (0.60, 0.70),
    "N": (0.84, 0.70),
    "M": (0.50, 0.43),
    "D": (0.50, 0.18),
}

labels = {
    "P": "P\nPacific\nsurface",
    "S": "S\nSouthern Ocean\nsurface",
    "L": "L\nLow latitude\nsurface",
    "N": "N\nNorth Atlantic\nsurface",
    "M": "M\nMid-depth",
    "D": "D\nDeep",
}

for b, (x, y) in pos.items():
    ax.text(x, y, labels[b], ha="center", va="center",
            bbox=dict(boxstyle="round,pad=0.45", fc="white", ec="black"))

# conceptual pathways
arrows = [("N", "M"), ("M", "D"), ("D", "S"), ("S", "P"), ("P", "L"), ("L", "N")]
for a, b in arrows:
    ax.annotate("", xy=pos[b], xytext=pos[a], arrowprops=dict(arrowstyle="->"))

# exchange-like connections
ax.annotate("", xy=pos["S"], xytext=pos["M"], arrowprops=dict(arrowstyle="<->", linestyle="--"))
ax.annotate("", xy=pos["L"], xytext=pos["M"], arrowprops=dict(arrowstyle="<->", linestyle="--"))

ax.set_xlim(0, 1)
ax.set_ylim(0, 0.95)
ax.axis("off")
ax.set_title("Toggweiler-style six-box model: PSLNMD")
plt.show()
````

## 6. P を分ける意味 / Why separate P?

P は極域表層を表します。  
P represents the Polar surface.

4-box では、低緯度表層をまとめて L として扱いました。  
In 4-box, low-latitude surface waters were grouped into L.

しかし、極域表層を分けると、次のような問いを扱えます。  
Separating the Polar surface allows us to ask:

```text
極域表層はどれくらい大気 CO2 と交換するか？
極域表層は南大洋や低緯度表層とどうつながるか？
太平洋側に蓄積した炭素は大気 CO2 にどう効くか？
```

```text
How strongly does the Polar surface exchange CO2 with the atmosphere?
How is the Polar surface connected to the Southern Ocean and low-latitude surface?
How does carbon stored on the Pacific side affect atmospheric CO2?
```

P は、海洋炭素循環を「大西洋中心」だけで見ないために重要です。  
P is important because it prevents us from viewing ocean carbon cycling only from an Atlantic-centered perspective.

## 7. M を分ける意味 / Why separate M?

M は中層を表します。  
M represents mid-depth water.

4-box では、内部海洋を D だけで表していました。  
In 4-box, the ocean interior was represented only by D.

しかし、M を分けると、次の問いが扱えます。  
Separating M allows us to ask:

```text
中層水は深層水より若いのか？
中層は表層とどれくらい交換するのか？
中層と深層では DIC, O2, Δ14C はどう違うのか？
```

```text
Is mid-depth water younger than deep water?
How strongly does M exchange with the surface?
How do DIC, O2, and Δ14C differ between M and D?
```

M を入れることで、内部海洋を単一の貯蔵庫ではなく、鉛直構造を持つ場所として扱えます。  
By adding M, the interior ocean is no longer a single reservoir, but a region with vertical structure.

## 8. 仮の体積を置く / Assign provisional volumes

ここでは、教育用に仮の体積を置きます。  
Here we assign provisional volumes for teaching.

厳密な値ではなく、箱の大きさの違いを理解するための値です。  
These are not exact values; they are for understanding differences in box size.

後で実際の Fortran/Python コードに合わせるときは、パラメタ値を置き換えます。  
Later, when matching the actual Fortran/Python code, we will replace these with the actual parameter values.

````python
VOCN_TOTAL = 1.292e18

VOLUME = {
    "P": 0.18 * VOCN_TOTAL,
    "S": 0.05 * VOCN_TOTAL,
    "L": 0.12 * VOCN_TOTAL,
    "N": 0.03 * VOCN_TOTAL,
    "M": 0.27 * VOCN_TOTAL,
}
VOLUME["D"] = VOCN_TOTAL - sum(VOLUME.values())

pd.DataFrame({
    "Box": boxes,
    "Volume [m3]": [VOLUME[b] for b in boxes],
    "Volume fraction": [VOLUME[b] / VOCN_TOTAL for b in boxes],
})
````

## 9. 単純な循環を置く / A simple circulation

まず、PSLNMD の直感をつかむために、単純な循環を置きます。  
First, to build intuition for PSLNMD, we set a simple circulation.

```text
N → M → D → S → P → L → N
```

これは、次のようなイメージです。  
This represents:

```text
N: North Atlantic sinking
M: mid-depth pathway
D: deep reservoir
S: Southern Ocean upwelling
P, L: surface return pathway
N: return to North Atlantic
```

この輸送は、厳密な論文モデルそのものではありません。  
This transport is not yet the exact model used in the paper.

ここでは、箱の意味を理解するための最小モデルとして使います。  
Here, it is used as a minimal model for understanding the meaning of the boxes.

````python
Q = 2.0e7
pathway = [("N", "M"), ("M", "D"), ("D", "S"), ("S", "P"), ("P", "L"), ("L", "N")]

def build_flux_matrix(pathway, Q, boxes):
    n = len(boxes)
    idx = {b: i for i, b in enumerate(boxes)}
    F = np.zeros((n, n))

    for source, target in pathway:
        i = idx[target]
        j = idx[source]
        F[i, j] += Q
        F[j, j] -= Q

    return F

F = build_flux_matrix(pathway, Q, boxes)
pd.DataFrame(F, index=boxes, columns=boxes)
````

## 10. 質量保存を確認する / Check mass conservation

閉じた循環では、各箱に入る流量と出る流量が等しくなります。  
In a closed circulation, each box has equal inflow and outflow.

これは、後で保存トレーサーを扱うときに重要です。  
This is important later when handling conserved tracers.

````python
rows = []
for b in boxes:
    inflow = sum(Q for a, bb in pathway if bb == b)
    outflow = sum(Q for aa, b2 in pathway if aa == b)
    rows.append({
        "Box": b,
        "Inflow [m3/s]": inflow,
        "Outflow [m3/s]": outflow,
        "Inflow - Outflow": inflow - outflow,
    })

pd.DataFrame(rows)
````

## 11. 染料実験: N に入れる / Dye experiment: release in N

N は北大西洋表層で、沈み込みの入口です。  
N is the North Atlantic surface box and the entrance of sinking.

N に染料を入れると、M, D へ入っていくはずです。  
If dye is released in N, it should enter M and D.

````python
volumes = np.array([VOLUME[b] for b in boxes])

def one_step(c, F):
    return c + (F @ c) * DT / volumes

def run_dye(source_box="N", years=1500):
    c = np.zeros(len(boxes))
    c[boxes.index(source_box)] = 1.0
    hist = []
    for day in range(years * 365 + 1):
        if day % 365 == 0:
            row = {"year": day / 365}
            for i, b in enumerate(boxes):
                row[b] = c[i]
            hist.append(row)
        c = one_step(c, F)
    return pd.DataFrame(hist)

dye_N = run_dye("N")

plt.figure()
for b in boxes:
    plt.plot(dye_N["year"], dye_N[b], label=b)
plt.xlabel("Time [yr]")
plt.ylabel("Dye concentration")
plt.title("Dye released in N: PSLNMD")
plt.legend()
plt.grid(True)
plt.show()
````

## 12. 染料実験: S に入れる / Dye experiment: release in S

S は南大洋表層です。  
S is the Subpolar surface box.

S に入れた染料は、表層経路 P, L, N を通り、その後 M, D に入ります。  
Dye released in S moves along the surface pathway P, L, N, and then enters M and D.

````python
dye_S = run_dye("S")

plt.figure()
for b in boxes:
    plt.plot(dye_S["year"], dye_S[b], label=b)
plt.xlabel("Time [yr]")
plt.ylabel("Dye concentration")
plt.title("Dye released in S: PSLNMD")
plt.legend()
plt.grid(True)
plt.show()
````

## 13. N source と S source を比較する / Compare N source and S source

N に入れた染料と S に入れた染料では、D への届き方が違います。  
Dye released in N and dye released in S reach D differently.

これは、N が沈み込みの入口であり、S が湧昇と表層経路の入口であるためです。  
This is because N is the entrance of sinking, while S is the entrance of upwelling and the surface pathway.

````python
plt.figure()
plt.plot(dye_N["year"], dye_N["D"], label="source N")
plt.plot(dye_S["year"], dye_S["D"], label="source S")
plt.xlabel("Time [yr]")
plt.ylabel("Dye in D")
plt.title("Arrival of dye to D")
plt.legend()
plt.grid(True)
plt.show()

pd.DataFrame({
    "Source": ["N", "S"],
    "D at 500 yr": [
        dye_N.loc[dye_N["year"] == 500, "D"].iloc[0],
        dye_S.loc[dye_S["year"] == 500, "D"].iloc[0],
    ],
    "D at 1500 yr": [dye_N["D"].iloc[-1], dye_S["D"].iloc[-1]],
})
````
