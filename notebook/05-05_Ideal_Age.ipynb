## 1. Ideal Age とは / What is Ideal Age?

Ideal Age は『最後に表層へ接した時から経過した時間』です。

Ideal Age is the elapsed time since a water parcel last contacted the surface.


```python
boxes=['P','S','L','N','M','D']
age=np.zeros(len(boxes))
history=[]
for y in range(3001):
    if y>0:
        age[4]+=1
        age[5]+=1
        age[4]*=0.9995
        age[5]=0.9998*age[5]+0.0002*age[4]
        age[:4]=0
    if y%10==0:
        history.append(age.copy())
history=np.array(history)
plt.plot(np.arange(history.shape[0])*10,history[:,4],label='M')
plt.plot(np.arange(history.shape[0])*10,history[:,5],label='D')
plt.xlabel('Year');plt.ylabel('Ideal age (yr)');plt.grid();plt.legend();plt.show()
```
