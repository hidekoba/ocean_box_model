```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Ideal Age とは / What is Ideal Age?

Ideal Age は『最後に表層へ接した時から経過した時間』です。

Ideal Age is the elapsed time since a water parcel last contacted the surface.


```python
boxes=['P','S','L','N','M','D']
age=np.zeros(len(boxes))
history=[]
for y in range(3001):
    if y>0:
        age[4]+=1
        age[5]+=1
        age[4]*=0.9995
        age[5]=0.9998*age[5]+0.0002*age[4]
        age[:4]=0
    if y%10==0:
        history.append(age.copy())
history=np.array(history)
plt.plot(np.arange(history.shape[0])*10,history[:,4],label='M')
plt.plot(np.arange(history.shape[0])*10,history[:,5],label='D')
plt.xlabel('Year');plt.ylabel('Ideal age (yr)');plt.grid();plt.legend();plt.show()
```

2. なぜ D は M より古くなるのか。 / Why does D become older than M?
3. 循環が強くなると Ideal Age はどう変わるか。 / How does stronger circulation affect Ideal Age?

### 想定解答 / Expected answers
- 表層は常に大気と接触するため年齢はリセットされる。 / Surface water is continuously reset by contact with the atmosphere.
- D は表層から最も隔離されるため最も古くなる。 / D is most isolated from the surface and therefore oldest.
- 循環が強いほどベンチレーションが速くなり Ideal Age は若くなる。 / Stronger circulation ventilates the ocean faster, reducing Ideal Age.
