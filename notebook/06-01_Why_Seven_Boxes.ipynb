## 今日の目的 / Goals

05 では、Toggweiler 型 6-box モデル **PSLNMD** を学びました。  
In 05, we learned the Toggweiler-style six-box model **PSLNMD**.

```text
P S L N M D
```

06 では、これを 7-box モデル **PSLNMAD** へ拡張します。  
In 06, we extend it to the seven-box model **PSLNMAD**.

```text
P S L N M A D
```

新しく追加される box は **A** です。  
The newly added box is **A**.

ここで A は大気ではありません。  
Here, A is **not** the atmosphere.

A は、北大西洋で沈み込んだ水が通る深層、つまり NADW 的な経路を表す box として考えます。  
A represents the deep pathway of water sinking in the North Atlantic, similar to a NADW pathway.

> **6-box の D を、A と D に分けることで、北大西洋起源深層水とより古い深層水を区別する。**  
> **By splitting the 6-box D into A and D, we distinguish North Atlantic-origin deep water from older deep water.**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8, 4.8)
````

## 1. Question: なぜ A box を追加するのか / Why add the A box?

6-box PSLNMD では、内部海洋は M と D の 2 つでした。  
In the six-box PSLNMD model, the ocean interior had two boxes: M and D.

```text
M: mid-depth
D: deep
```

しかし、深層 D の中にも、異なる性質の水があります。  
However, even inside the deep ocean D, there are waters with different properties.

特に重要なのは、北大西洋で沈み込んだ水です。  
Especially important is water that sinks in the North Atlantic.

この水は、深層へ入りますが、最も古い深層水と完全に同じではありません。  
This water enters the deep ocean, but it is not identical to the oldest deep water.

そこで、7-box では **A** を追加します。  
Therefore, in the seven-box model we add **A**.

```text
N → A → D
```

A は、北大西洋沈み込みが通る深層経路を表します。  
A represents the deep pathway through which North Atlantic sinking passes.

## 2. 6-box と 7-box の違い / Difference between six-box and seven-box

6-box:

```text
P S L N M D
```

7-box:

```text
P S L N M A D
```

違いは、6-box の深層 D を、7-box では A と D に分けることです。  
The difference is that the deep box D in the six-box model is divided into A and D in the seven-box model.

```text
6-box:
N → M → D

7-box:
N → A → D
M remains as mid-depth water
```

つまり、7-box では、

```text
M: 中層
A: 北大西洋起源深層水の経路
D: より古い深層水
```

を区別します。

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

pos6 = {"P": (0.12,0.70), "S": (0.36,0.70), "L": (0.60,0.70), "N": (0.84,0.70),
        "M": (0.50,0.43), "D": (0.50,0.18)}
path6 = [("N","M"),("M","D"),("D","S"),("S","P"),("P","L"),("L","N")]
for b,(x,y) in pos6.items():
    axes[0].text(x,y,b,ha="center",va="center",bbox=dict(boxstyle="round,pad=0.5",fc="white",ec="black"))
for a,b in path6:
    axes[0].annotate("", xy=pos6[b], xytext=pos6[a], arrowprops=dict(arrowstyle="->"))
axes[0].set_title("6-box: PSLNMD")
axes[0].set_xlim(0,1); axes[0].set_ylim(0,1); axes[0].axis("off")

pos7 = {"P": (0.12,0.72), "S": (0.36,0.72), "L": (0.60,0.72), "N": (0.84,0.72),
        "M": (0.34,0.43), "A": (0.66,0.43), "D": (0.50,0.18)}
path7 = [("N","A"),("A","D"),("D","S"),("S","P"),("P","L"),("L","N")]
for b,(x,y) in pos7.items():
    axes[1].text(x,y,b,ha="center",va="center",bbox=dict(boxstyle="round,pad=0.5",fc="white",ec="black"))
for a,b in path7:
    axes[1].annotate("", xy=pos7[b], xytext=pos7[a], arrowprops=dict(arrowstyle="->"))
axes[1].annotate("", xy=pos7["M"], xytext=pos7["S"], arrowprops=dict(arrowstyle="<->", linestyle="--"))
axes[1].annotate("", xy=pos7["M"], xytext=pos7["L"], arrowprops=dict(arrowstyle="<->", linestyle="--"))
axes[1].set_title("7-box: PSLNMAD")
axes[1].set_xlim(0,1); axes[1].set_ylim(0,1); axes[1].axis("off")
plt.show()
````

## 6. 最小輸送経路 / Minimal transport pathway

7-box の最小経路として、次を使います。  
As a minimal seven-box pathway, we use:

```text
N → A → D → S → P → L → N
```

M は、表層 S, L と交換する中層として扱います。  
M is treated as mid-depth water exchanging with surface boxes S and L.

```text
S ↔ M
L ↔ M
```

````python
Q = 2.0e7
pathway = [("N", "A"), ("A", "D"), ("D", "S"), ("S", "P"), ("P", "L"), ("L", "N")]
exchanges = [("S", "M", 0.3e7), ("L", "M", 0.2e7)]
pd.DataFrame(pathway, columns=["From", "To"])
````

````python
fig, ax = plt.subplots(figsize=(8, 5))
pos = {"P": (0.12,0.72), "S": (0.36,0.72), "L": (0.60,0.72), "N": (0.84,0.72),
       "M": (0.34,0.43), "A": (0.66,0.43), "D": (0.50,0.18)}

for b,(x,y) in pos.items():
    ax.text(x,y,b,ha="center",va="center",bbox=dict(boxstyle="round,pad=0.55",fc="white",ec="black"))
for a,b in pathway:
    ax.annotate("", xy=pos[b], xytext=pos[a], arrowprops=dict(arrowstyle="->"))
for a,b,q in exchanges:
    ax.annotate("", xy=pos[b], xytext=pos[a], arrowprops=dict(arrowstyle="<->", linestyle="--"))

ax.set_xlim(0,1); ax.set_ylim(0,1); ax.axis("off")
ax.set_title("Minimal PSLNMAD pathway")
plt.show()
````

## 7. 輸送行列を作る / Build the transport matrix

6-box model で学んだ方法をそのまま使って、7×7 の輸送行列を作ります。  
Using the same method learned in 6-box model, we build a 7×7 transport matrix.

````python
def build_flux_matrix(pathway, Q, boxes):
    F = np.zeros((len(boxes), len(boxes)))
    idx = {b: i for i, b in enumerate(boxes)}
    for source, target in pathway:
        i = idx[target]
        j = idx[source]
        F[i, j] += Q
        F[j, j] -= Q
    return F

def build_flux_matrix_with_exchange(pathway, Q, boxes, exchanges=None):
    F = build_flux_matrix(pathway, Q, boxes)
    idx = {b: i for i, b in enumerate(boxes)}
    if exchanges is None:
        exchanges = []
    for a, b, q in exchanges:
        ia = idx[a]
        ib = idx[b]
        F[ib, ia] += q
        F[ia, ia] -= q
        F[ia, ib] += q
        F[ib, ib] -= q
    return F

F = build_flux_matrix_with_exchange(pathway, Q, boxes, exchanges)
pd.DataFrame(F, index=boxes, columns=boxes)
````

## 8. 保存チェック / Conservation check

7-box でも、保存トレーサーの総量が保存されることを確認します。  
Even in the seven-box model, we check that the total amount of a passive tracer is conserved.

````python
volumes = np.array([VOLUME[b] for b in boxes])
idx = {b: i for i, b in enumerate(boxes)}
DT = 8.64e4

def one_step_transport(c, F):
    return c + (F @ c) * DT / volumes

def total_amount(c):
    return np.sum(c * volumes)

def run_dye(source_box="N", years=1000):
    c = np.zeros(len(boxes))
    c[idx[source_box]] = 1.0
    hist = []
    for day in range(years * 365 + 1):
        if day % 365 == 0:
            row = {"year": day/365, "total": total_amount(c)}
            for i,b in enumerate(boxes):
                row[b] = c[i]
            hist.append(row)
        c = one_step_transport(c, F)
    return pd.DataFrame(hist)

h = run_dye("N", years=1000)
print("Relative change:", (h["total"].iloc[-1]-h["total"].iloc[0])/h["total"].iloc[0])

plt.figure()
plt.plot(h["year"], h["total"])
plt.xlabel("Time [yr]")
plt.ylabel("Total dye amount")
plt.title("Conservation check")
plt.grid(True)
plt.show()
````

## 9. N に染料を入れる / Dye released in N

N で沈み込んだ水は、まず A に入り、その後 D へ進むはずです。  
Water sinking in N should first enter A and then proceed to D.

````python
dye_N = run_dye("N", years=1500)

plt.figure()
for b in boxes:
    plt.plot(dye_N["year"], dye_N[b], label=b)
plt.xlabel("Time [yr]")
plt.ylabel("Dye concentration")
plt.title("Dye released in N: PSLNMAD")
plt.legend()
plt.grid(True)
plt.show()
````
