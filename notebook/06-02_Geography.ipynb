## 1. Question: PSLNMAD はどこにあるのか / Where are PSLNMAD boxes?

7-box モデルでは、表層 box が 4 つあります。  
In the seven-box model, there are four surface boxes.

```text
P: Pacific surface
S: Polar surface
L: Low-latitude surface
N: North Atlantic surface
```

内部 box は 3 つです。  
There are three interior boxes.

```text
M: mid-depth ocean
A: Atlantic deep pathway
D: deep ocean
```

6-box との大きな違いは、深層を A と D に分けることです。  
The main difference from the six-box model is that the deep ocean is divided into A and D.

## 4. 鉛直構造 / Vertical structure

In the 7-box model, the interior ocean is divided into three boxes.

```text
M: mid-depth
A: Atlantic deep pathway
D: deep ocean
```

M は中層です。  
M is mid-depth water.

A は北大西洋起源深層水が通る経路です。  
A is the pathway of North Atlantic-origin deep water.

D はより古い深層水です。

## 5. A box の意味 / Meaning of the A box

A は 7-box model の核心です。  
A is the core new feature of the 7-box model.

A は **Atlantic deep pathway** と考えます。  
We interpret A as the **Atlantic deep pathway**.

```text
N → A → D
```

この経路により、北大西洋で沈み込んだ水がすぐに古い D と完全混合するのではなく、A という経路を通ることを表現できます。  
This pathway represents that water sinking in the North Atlantic does not immediately become fully mixed with old D water, but passes through A.

この分離により、次を考えやすくなります。  
This separation helps us examine:

```text
A と D の年齢差
A と D の O2 差
A と D の DIC 差
A と D の Δ14C 差
```

```text
age difference between A and D
O2 difference between A and D
DIC difference between A and D
Δ14C difference between A and D
```

## 6. 仮の体積と代表深度 / Provisional volumes and representative depths

ここでは、教育用の仮の体積と代表深度を設定します。  
Here we assign provisional volumes and representative depths for teaching.

````python
VOCN_TOTAL = 1.292e18

VOLUME = {
    "P": 0.18 * VOCN_TOTAL,
    "S": 0.05 * VOCN_TOTAL,
    "L": 0.12 * VOCN_TOTAL,
    "N": 0.03 * VOCN_TOTAL,
    "M": 0.27 * VOCN_TOTAL,
    "A": 0.12 * VOCN_TOTAL,
}
VOLUME["D"] = VOCN_TOTAL - sum(VOLUME.values())

DEPTH = {
    "P": 100,
    "S": 100,
    "L": 100,
    "N": 100,
    "M": 1000,
    "A": 2500,
    "D": 4000,
}

geom = pd.DataFrame({
    "Box": boxes,
    "Volume fraction": [VOLUME[b] / VOCN_TOTAL for b in boxes],
    "Representative depth [m]": [DEPTH[b] for b in boxes],
})
geom
````

````python
plt.figure()
plt.bar(geom["Box"], geom["Volume fraction"])
plt.xlabel("Box")
plt.ylabel("Volume fraction")
plt.title("Provisional volume fractions")
plt.grid(True, axis="y", alpha=0.3)
plt.show()

plt.figure()
plt.bar(geom["Box"], geom["Representative depth [m]"])
plt.xlabel("Box")
plt.ylabel("Representative depth [m]")
plt.title("Representative depth")
plt.gca().invert_yaxis()
plt.grid(True, axis="y", alpha=0.3)
plt.show()
````

## 7. 地理・鉛直構造・輸送をつなげる / Connect geography, vertical structure, and transport

PSLNMAD の最小輸送経路は、

```text
N → A → D → S → P → L → N
```

です。

さらに中層 M は、

```text
S ↔ M
L ↔ M
```

で表層と交換します。

この構造により、

```text
N で沈み込む新しい深層水
A を通る北大西洋起源深層水
D にあるより古い水
S で表層へ戻る水
```

を分けて考えられます。  
This structure separates North Atlantic sinking water, Atlantic deep pathway water in A, older water in D, and upwelling water in S.

````python
pathway = [("N", "A"), ("A", "D"), ("D", "S"), ("S", "P"), ("P", "L"), ("L", "N")]
exchanges = [("S", "M", 0.3e7), ("L", "M", 0.2e7)]

pd.DataFrame(pathway, columns=["From", "To"])
````

## 8. 簡単な年齢予想 / Simple age prediction

A と D を分けたとき、どちらが古くなるでしょうか。  
When A and D are separated, which becomes older?

直感的には、

```text
A: N から入ってくる比較的新しい深層水
D: さらに下流にある古い深層水
```

なので、D の方が古くなりやすいと予想されます。  
Intuitively, A receives relatively young deep water from N, while D is farther downstream, so D tends to be older.

````python
# very simple ideal age demonstration
Q = 2.0e7
DT = 8.64e4
SEC_PER_YEAR = 365 * 86400
DT_YEAR = DT / SEC_PER_YEAR
volumes = np.array([VOLUME[b] for b in boxes])
idx = {b: i for i, b in enumerate(boxes)}

def build_flux_matrix(pathway, Q, boxes):
    F = np.zeros((len(boxes), len(boxes)))
    for source, target in pathway:
        i = idx[target]
        j = idx[source]
        F[i, j] += Q
        F[j, j] -= Q
    return F

def add_exchange(F, a, b, q):
    ia, ib = idx[a], idx[b]
    F[ib, ia] += q
    F[ia, ia] -= q
    F[ia, ib] += q
    F[ib, ib] -= q
    return F

F = build_flux_matrix(pathway, Q, boxes)
for a, b, q in exchanges:
    F = add_exchange(F, a, b, q)

def one_step_age(age):
    new = age + (F @ age) * DT / volumes
    for b in ["M", "A", "D"]:
        new[idx[b]] += DT_YEAR
    for b in ["P", "S", "L", "N"]:
        new[idx[b]] = 0.0
    return new

age = np.zeros(len(boxes))
hist = []
for day in range(3000 * 365 + 1):
    if day % 365 == 0:
        row = {"year": day / 365}
        for i, b in enumerate(boxes):
            row[b] = age[i]
        hist.append(row)
    age = one_step_age(age)

age_df = pd.DataFrame(hist)

plt.figure()
for b in ["M", "A", "D"]:
    plt.plot(age_df["year"], age_df[b], label=b)
plt.xlabel("Time [yr]")
plt.ylabel("Ideal age [yr]")
plt.title("Simple ideal age in PSLNMAD")
plt.legend()
plt.grid(True)
plt.show()

age_df.tail()
````
