## Goals / 今日の目標
- 7×7 輸送行列を構築する / Build the 7×7 transport matrix.
- A box を含む循環を Python で実装する / Implement circulation including the A box.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

## 1. Transport pathway

```text
N → A → D → S → P → L → N
S ↔ M
L ↔ M
```


````python
boxes=['P','S','L','N','M','A','D']
pathway=[('N','A'),('A','D'),('D','S'),('S','P'),('P','L'),('L','N')]
exchange=[('S','M',0.3e7),('L','M',0.2e7)]
Q=2e7
idx={b:i for i,b in enumerate(boxes)}
F=np.zeros((7,7))
for s,t in pathway:
 i=idx[t];j=idx[s];F[i,j]+=Q;F[j,j]-=Q
for a,b,q in exchange:
 ia,ib=idx[a],idx[b]
 F[ib,ia]+=q;F[ia,ia]-=q
 F[ia,ib]+=q;F[ib,ib]-=q
pd.DataFrame(F,index=boxes,columns=boxes)
````

## 2. 行列を読む / Reading the matrix

列は流出元、行は流入先です。対角成分は流出量を表します。

Columns are sources, rows are destinations, and diagonal terms represent outflow.


````python
plt.figure(figsize=(6,5))
plt.imshow(F,cmap='coolwarm')
plt.xticks(range(7),boxes)
plt.yticks(range(7),boxes)
plt.colorbar(label='Transport')
plt.title('7x7 Transport Matrix')
plt.show()
````

## 3. 保存確認 / Conservation


````python
print('Column sums')
print(pd.Series(F.sum(axis=0),index=boxes))
````

保存トレーサーでは各列和が 0 になることが重要です。

For conservative transport, each column sum should be zero.
