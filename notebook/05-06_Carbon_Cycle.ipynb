```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. 保存トレーサーと非保存トレーサー / Conservative vs. non-conservative tracers

保存トレーサーは輸送だけで変化します。/ Conservative tracers change only by transport.

炭素や酸素は、生物活動や大気海洋交換によっても変化します。/ Carbon and oxygen also change because of biology and air-sea exchange.


## 2. Biological pump / 生物ポンプ

最も単純には、表層で DIC が減少し、内部で DIC が増加します。

In the simplest picture, DIC decreases at the surface and increases in the interior.


```python
boxes=['P','S','L','N','M','D']
DIC=np.array([2100,2100,2100,2100,2200,2300],dtype=float)
PO4=np.array([0.5,0.6,0.4,0.3,1.5,2.3])
O2=np.array([260,280,250,300,180,120])
df=pd.DataFrame({'Box':boxes,'DIC':DIC,'PO4':PO4,'O2':O2})
df
```

```python
plt.figure(figsize=(8,4))
plt.plot(boxes,DIC,'o-',label='DIC')
plt.ylabel('µmol kg$^{-1}$')
plt.grid();plt.legend();plt.show()
```

## 3. 簡単な再無機化 / Simple remineralization

ここでは教育用として、表層から M と D へ炭素を移動させます。

For teaching purposes we transfer carbon from the surface to M and D.


```python
bio=10.0
DIC2=DIC.copy()
DIC2[:4]-=bio
DIC2[4]+=bio*0.4
DIC2[5]+=bio*0.6
pd.DataFrame({'Box':boxes,'Before':DIC,'After':DIC2})
```

```python
plt.figure(figsize=(8,4))
x=np.arange(len(boxes))
plt.bar(x-0.2,DIC,0.4,label='Before')
plt.bar(x+0.2,DIC2,0.4,label='After')
plt.xticks(x,boxes)
plt.ylabel('DIC')
plt.legend();plt.grid(axis='y');plt.show()
```

## 4. Redfield 比 / Redfield ratio

PO4 が増えると DIC も増え、O2 は減少します。

As PO4 increases, DIC increases while O2 decreases.


```python
redfield_C=106
redfield_O2=-138
delta_po4=0.2
print('ΔDIC =',redfield_C*delta_po4)
print('ΔO2  =',redfield_O2*delta_po4)
```
