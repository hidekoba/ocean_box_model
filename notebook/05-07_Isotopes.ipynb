```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. なぜ同位体を使うのか / Why do we use isotopes?

濃度だけでは、水塊の履歴は分かりません。/ Concentrations alone do not reveal water-mass history.

δ13C は生物ポンプを、Δ14C はベンチレーション年代を調べるために利用されます。
δ13C reflects the biological pump, whereas Δ14C is widely used as a ventilation-age tracer.


```python
boxes=['P','S','L','N','M','D']
delta13C=np.array([2.0,1.8,1.9,2.1,0.8,0.2])
Delta14C=np.array([0,-20,-10,10,-120,-220])
pd.DataFrame({'Box':boxes,'δ13C (‰)':delta13C,'Δ14C (‰)':Delta14C})
```

```python
plt.figure(figsize=(8,4))
plt.plot(boxes,delta13C,'o-',label='δ13C')
plt.ylabel('‰')
plt.grid();plt.legend();plt.show()
plt.figure(figsize=(8,4))
plt.plot(boxes,Delta14C,'s-',label='Δ14C')
plt.ylabel('‰')
plt.grid();plt.legend();plt.show()
```

## 2. δ13C と生物ポンプ / δ13C and the biological pump

植物プランクトンは 12C を選択的に取り込むため、表層の δ13C は高くなり、再無機化が起こる深層では低くなります。

Phytoplankton preferentially assimilate 12C, leaving high δ13C at the surface and lower δ13C in deep water after remineralization.


## 3. Δ14C とベンチレーション / Δ14C and ventilation

14C は放射壊変するため、表層から隔離された時間が長いほど Δ14C は小さくなります。

Because 14C decays radioactively, Δ14C decreases as water remains isolated from the atmosphere.


```python
age=np.array([0,0,0,0,500,1500])
tau=8267.0
model=np.exp(-age/tau)
plt.figure(figsize=(6,4))
plt.plot(age,model,'o-')
plt.xlabel('Ideal age (yr)')
plt.ylabel('Relative 14C')
plt.grid();plt.show()
```

## 4. Ideal Age と Δ14C / Linking Ideal Age and Δ14C

05-05 で学んだ Ideal Age が大きいほど、Δ14C は低くなる傾向があります。

Larger Ideal Age generally corresponds to lower Δ14C.
