## 1. 7-box モデルの構造 / 7-box structure

```
P  S  L  N
 \ | | /
    M
    |
    A
    |
    D
```

A は北大西洋起源深層水、D はより古い深層水です。
A is the Atlantic deep pathway, while D is older deep water.

## 2. それぞれのトレーサーが教えてくれること / What each tracer tells us

|Tracer|Meaning|
|---|---|
|Passive tracer|transport|
|Ideal Age|ventilation age|
|DIC|carbon storage|
|PO4|nutrients|
|O2|remineralization|
|δ13C|biology + mixing|
|Δ14C|ventilation + isolation|

````python
summary=pd.DataFrame({
"Tracer":["Passive","Ideal Age","DIC","PO4","O2","δ13C","Δ14C"],
"Main information":[
"Transport",
"Ventilation age",
"Carbon storage",
"Nutrients",
"Remineralization",
"Biological pump + mixing",
"Ventilation + isolation time"]
})
summary
````

## 3. 概念図 / Concept map

```
Circulation
     ↓
Ventilation
     ↓
Ideal Age
     ↓
Δ14C

Biological pump
     ↓
PO4
     ↓
DIC
     ↓
δ13C
     ↓
O2
```

````python
labels=["Passive","Age","DIC","PO4","O2","δ13C","Δ14C"]
scores=[5,5,4,4,4,5,5]
plt.figure(figsize=(8,4))
plt.bar(labels,scores)
plt.ylabel("Importance (teaching)")
plt.grid(axis="y",alpha=0.3)
plt.show()
````

## 4. Integrated discussion

- A は若い北大西洋起源深層水  
  A is relatively young North Atlantic deep water.

- D は古い深層水  
  D is older deep water.

- ベンチレーションが弱いほど Ideal Age は増加する。  
  Weaker ventilation increases Ideal Age.

- Δ14C は低下する。  
  Δ14C decreases.

- DIC は蓄積しやすい。  
  DIC accumulates.

- O2 は減少しやすい。  
  O2 decreases.
