## 9. Traceback の読み方

一番最後の行を見ることが重要です。

例

```text
NameError: name 'tempperature' is not defined
```

まず最後の行、その次にエラーが起きた行番号を確認します。

## 10. Debugging workflow

1. エラーメッセージ最後の行を読む
2. 行番号を確認する
3. 変数名・括弧・インデントを確認する
4. 小さく修正する
5. もう一度実行する
