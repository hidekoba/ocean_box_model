```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Dataset と DataArray

```python
depth=np.array([0,100,500,1000])
temp=np.array([20,15,8,4])

da=xr.DataArray(temp,
                coords={"depth":depth},
                dims=["depth"],
                name="temperature")
da
```

## 2. Dataset

```python
ds=xr.Dataset({"temperature":da})
ds
```

## 3. Coordinates

```python
print(ds.coords)
print(ds.depth.values)
```

## 4. Selection

```python
print(ds.temperature.sel(depth=500))
print(ds.temperature.isel(depth=2))
```

## 5. Statistics

```python
print(ds.temperature.mean())
print(ds.temperature.max())
print(ds.temperature.min())
```

## 6. Plot

```python
ds.temperature.plot(marker="o")
plt.gca().invert_yaxis()
plt.show()
```

## 7. Write NetCDF

```python
ds.to_netcdf("sample.nc")
```

## 8. Read NetCDF

```python
ds2=xr.open_dataset("sample.nc")
ds2
```
