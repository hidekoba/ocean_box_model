```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Series

```python
s=pd.Series([20,18,15,10,5],name="Temperature")
print(s)
```

## 2. DataFrame

```python
df=pd.DataFrame({
"Depth":[0,100,500,1000,2000],
"Temperature":[20,18,15,10,5],
"Salinity":[34.0,34.2,34.5,34.7,34.8]
})
df
```

## 3. Columns

```python
print(df["Temperature"])
print(df[["Depth","Salinity"]])
```

## 4. Statistics

```python
print(df.describe())
```

## 5. Selecting rows

```python
print(df.iloc[0])
print(df.iloc[-1])
print(df[df["Depth"]>=500])
```

## 6. Add a column

```python
df["Temperature_K"]=df["Temperature"]+273.15
df
```

## 7. Save and read CSV

```python
df.to_csv("ocean_profile.csv",index=False)
df2=pd.read_csv("ocean_profile.csv")
df2
```

## 8. Plot from DataFrame

```python
plt.figure(figsize=(5,6))
plt.plot(df["Temperature"],df["Depth"],"o-")
plt.gca().invert_yaxis()
plt.xlabel("Temperature [°C]")
plt.ylabel("Depth [m]")
plt.grid(True)
plt.show()
```
