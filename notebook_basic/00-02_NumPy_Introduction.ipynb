```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. ndarray を作る

```python
a=np.array([1,2,3,4,5])
print(a)
print(type(a))
```

## 2. 配列演算

```python
x=np.arange(5)
print(x)
print(x+1)
print(x*2)
print(x**2)
```

## 3. indexing / slicing

```python
x=np.arange(10)
print(x[0])
print(x[-1])
print(x[2:6])
```

## 4. linspace

```python
depth=np.linspace(0,5000,11)
print(depth)
```

## 5. reshape

```python
a=np.arange(12)
b=a.reshape(3,4)
print(b)
```

## 6. Statistics

```python
temp=np.array([20,18,15,10,5])
print(temp.mean())
print(temp.min())
print(temp.max())
print(temp.std())
```

## 7. meshgrid

```python
x=np.linspace(0,10,6)
y=np.linspace(0,5,4)
X,Y=np.meshgrid(x,y)
print(X.shape)
print(Y.shape)
```

## 8. Plotting arrays

```python
x=np.linspace(0,2*np.pi,200)
y=np.sin(x)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

## 9. Ocean example

```python
depth=np.array([0,100,500,1000,2000,4000])
temp=np.array([20,16,8,4,2.5,1.5])

plt.figure(figsize=(5,6))
plt.plot(temp,depth,'o-')
plt.gca().invert_yaxis()
plt.xlabel("Temperature")
plt.ylabel("Depth")
plt.grid(True)
plt.show()
```
