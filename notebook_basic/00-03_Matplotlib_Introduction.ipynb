```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Line plot

```python
x=np.linspace(0,2*np.pi,200)
plt.plot(x,np.sin(x))
plt.grid(True)
plt.show()
```

## 2. Labels and title

```python
plt.plot(x,np.cos(x))
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("Cosine")
plt.grid(True)
plt.show()
```

## 3. Scatter plot

```python
y=np.sin(x)
plt.scatter(x[::10],y[::10])
plt.grid(True)
plt.show()
```

## 4. Bar chart

```python
boxes=["P","S","L","N","M","A","D"]
vol=[18,5,12,3,27,12,23]
plt.bar(boxes,vol)
plt.ylabel("Relative volume")
plt.show()
```

## 5. Histogram

```python
data=np.random.normal(size=1000)
plt.hist(data,bins=30)
plt.show()
```

## 6. Ocean profile

```python
depth=np.array([0,100,500,1000,2000,4000])
temp=np.array([20,16,8,4,2.5,1.5])

plt.figure(figsize=(5,6))
plt.plot(temp,depth,"o-",label="Temperature")
plt.gca().invert_yaxis()
plt.xlabel("Temperature [°C]")
plt.ylabel("Depth [m]")
plt.legend()
plt.grid(True)
plt.show()
```
