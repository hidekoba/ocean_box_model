## 1. パッケージ管理

推奨:

```bash
micromamba create -n geo python=3.12
micromamba activate geo
```

他の選択肢:

- conda
- pip

## 2. Python パッケージ

```bash
micromamba install numpy scipy matplotlib pandas xarray netcdf4
pip install jupyterlab
```

## 3. NetCDF ツール

```bash
ncdump -h file.nc
cdo sinfo file.nc
ncks -m file.nc
```

## 4. GPU

```bash
nvidia-smi
python -c "import torch;print(torch.cuda.is_available())"
```

## 5. SSH

```bash
ssh username@server
scp file.nc username@server:~
rsync -av data/ server:data/
```

## 6. tmux

```bash
tmux
Ctrl-b d
tmux attach
```

長時間ジョブに便利です。
