## 4. Figures

Every figure should have

- title
- x label
- y label
- units
- grid (if appropriate)

````python
import numpy as np
import matplotlib.pyplot as plt

depth=np.array([0,100,500,1000,2000,4000])
temp=np.array([20,16,8,4,2.5,1.5])

plt.figure(figsize=(5,6))
plt.plot(temp,depth,"o-")
plt.gca().invert_yaxis()
plt.xlabel("Temperature [°C]")
plt.ylabel("Depth [m]")
plt.title("Temperature profile")
plt.grid(True)
plt.show()
````
