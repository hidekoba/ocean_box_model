## 1. Recommended project structure

```text
project/
├── data/
├── notebook/
├── notebook-solutions/
├── python/
├── figures/
├── output/
├── README.md
└── environment.yml
```

## 2. Typical workflow

```text
Download data
    ↓
Linux
    ↓
Python / xarray
    ↓
Analysis
    ↓
Figures
    ↓
Notebook
    ↓
GitHub
```

## 3. Virtual environments

```bash
micromamba create -n ocean python=3.12
micromamba activate ocean
```

パッケージは環境ごとに管理します。

## 4. Version control

```bash
git status
git add .
git commit -m "Update notebook"
git push
```

## 5. Data management

- 元データは変更しない
- 出力は output/ に保存
- 図は figures/ に保存
- Notebook は notebook/ に保存

## 6. Backup strategy

- GitHub
- External SSD
- University server

最低 2 箇所以上へ保存しましょう。
