# Linux 基礎（科学計算入門）

## Goals / 目的

Linux の基本操作を学び、研究室で科学計算を始める準備をします。

Topics

- Terminal
- pwd

- man

## 1. Linux とは？

Linux はスーパーコンピュータや研究用サーバーで最も広く利用される OS です。

Linux is the most widely used operating system on research servers and supercomputers.

## 2. Terminal

Terminal はコマンドを入力する画面です。

基本形

```bash
command options arguments
```

## 3. 現在の場所

```bash
pwd
```

現在のディレクトリを表示します。

## 4. ファイル一覧

```bash
ls
ls -l
ls -lh
ls -a
```

## 5. ディレクトリ移動

```bash
cd
cd ..
cd ~
```

## 6. ディレクトリ作成

```bash
mkdir practice
cd practice
pwd
```

## 7. コピー・移動・削除

```bash
cp file1 file2
mv old new
rm file
rm -r directory
```

**注意:** `rm` は元に戻せません。

## 8. Tab 補完

長いファイル名は `Tab` キーで補完できます。

## 9. 履歴

```bash
history
```

以前実行したコマンドを確認できます。

## 10. ヘルプ

```bash
man ls
ls --help
```
