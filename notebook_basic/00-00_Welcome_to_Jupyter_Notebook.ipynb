## 1. Jupyter Notebook とは？ / What is Jupyter Notebook?

Notebook は複数の **セル (cell)** からできています。  
A notebook consists of multiple **cells**.

| Cell type | Purpose |
|---|---|
| Markdown cell | 説明、見出し、数式、表、考察 / Explanations, headings, equations, tables, discussion |
| Code cell | Python コードの実行 / Running Python code |

このセルは Markdown cell です。次のセルは Code cell です。  
This is a Markdown cell. The next cell is a code cell.

````python
print("Hello, Jupyter Notebook!")
print("こんにちは、Jupyter Notebook!")
````

## 2. セルを実行する / Run a cell

| Action | Shortcut |
|---|---|
| Run and move to next cell | `Shift + Enter` |
| Run and stay | `Ctrl + Enter` |
| Run and insert a new cell below | `Alt + Enter` |

次のセルを `Shift + Enter` で実行してください。  
Run the next cell with `Shift + Enter`.

````python
2 + 3
````

Code cell の最後の値は自動表示されます。複数の値には `print()` を使います。  
The last value in a code cell is displayed automatically. Use `print()` for multiple values.

````python
print(2 + 3)
print(10 / 4)
print(2 ** 8)
````

## 3. Python を計算機として使う / Use Python as a calculator

| Operation | Symbol |
|---|---|
| Addition | `+` |
| Subtraction | `-` |
| Multiplication | `*` |
| Division | `/` |
| Floor division | `//` |
| Remainder | `%` |
| Power | `**` |

````python
a = 12
b = 5

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
````

## 5. 実行順序 / Execution order

Notebook の状態は、画面上の順序ではなく **実際にセルを実行した順番**で決まります。  
Notebook state is determined by the **order cells were executed**, not only by their visual order.

````python
x = 10
print("x =", x)
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

## 10. 最初の配列 / Your first array

````python
depth = np.array([0, 100, 500, 1000, 2000, 4000])
temperature = np.array([20.0, 16.0, 8.0, 4.0, 2.5, 1.5])

print(depth)
print(temperature)
````

````python
temperature_kelvin = temperature + 273.15
temperature_kelvin
````

## 11. 最初の図 / Your first plot

````python
plt.figure(figsize=(5, 6))
plt.plot(temperature, depth, marker="o")
plt.gca().invert_yaxis()
plt.xlabel("Temperature [°C]")
plt.ylabel("Depth [m]")
plt.title("Ocean temperature profile")
plt.grid(True)
plt.show()
````

## 12. エラーは情報 / Errors are information

プログラミングではエラーが出るのが普通です。  
Errors are a normal part of programming.

- `NameError`: 未定義の変数 / Undefined variable
- `SyntaxError`: 文法ミス / Invalid syntax
- `IndexError`: 配列範囲外 / Index outside an array
- `ModuleNotFoundError`: ライブラリが未導入 / Missing library
- `IndentationError`: インデントの問題 / Indentation problem
