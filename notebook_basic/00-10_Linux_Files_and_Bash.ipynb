- wildcard (*)
- pipes (|)
- redirect (>)

## 1. ファイル作成

```bash
touch test.txt
```

## 2. ファイル表示

```bash
cat file.txt
less file.txt
head file.txt
tail file.txt
```

## 3. コピー・移動

```bash
cp file1 file2
mv old.txt new.txt
```

## 4. 削除

```bash
rm file.txt
rm -r directory
```

`rm` は元に戻せません。

## 5. 検索

```bash
find . -name "*.nc"
grep temperature file.txt
grep -r DIC .
```

## 6. パイプ

```bash
cat file.txt | grep temperature
ls | wc -l
```

## 7. リダイレクト

```bash
echo hello > test.txt
cat test.txt
```

## 8. ワイルドカード

```bash
ls *.nc
cp *.txt backup/
```

## Hands-on

```bash
mkdir practice
cd practice
touch a.txt b.txt
cp a.txt c.txt
ls
echo ocean > a.txt
cat a.txt
```

## Research examples

```bash
find . -name "*.nc"
grep latitude *.txt
du -sh .
df -h
```
